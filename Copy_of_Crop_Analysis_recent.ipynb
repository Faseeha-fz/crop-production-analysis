##**1. To analyze yearly trends in crop production, area, and productivity.**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import plotly.graph_objects as go
from sklearn.preprocessing import MinMaxScaler

# List of crops
crops = ['Paddy', 'Pepper', 'Turmeric', 'Arecanut', 'Banana', 'Betel leaves', 'Cardamom',
         'Cashew', 'Cloves', 'Cocoa', 'Coconut', 'Coffee', 'Ginger', 'Jack', 'Mango',
         'Nutmeg', 'Pappaya', 'Pineapple', 'Plantain', 'Rubber', 'Tapioca', 'Tea']

# Function to normalize the data
def normalize(df, columns):
    scaler = MinMaxScaler()
    df[columns] = scaler.fit_transform(df[columns])
    return df

# Iterate through each crop in the list
for crop in crops:
    # Load the dataset corresponding to the crop
    df = globals().get(crop, None)  # Use globals() to get the DataFrame with the crop name

    if df is not None and 'Agriculture Year' in df.columns and 'Production' in df.columns and 'Productivity' in df.columns and 'Area' in df.columns:
        # Normalize the columns
        df_normalized = normalize(df.copy(), ['Production', 'Productivity', 'Area'])

        # Create a Plotly figure
        fig = go.Figure()

        # Add Production trace
        fig.add_trace(go.Scatter(x=df_normalized['Agriculture Year'], y=df_normalized['Production'],
                                 mode='lines+markers', name='Production', line=dict(color='blue')))

        # Add Productivity trace
        fig.add_trace(go.Scatter(x=df_normalized['Agriculture Year'], y=df_normalized['Productivity'],
                                 mode='lines+markers', name='Productivity', line=dict(color='green')))

        # Add Area trace
        fig.add_trace(go.Scatter(x=df_normalized['Agriculture Year'], y=df_normalized['Area'],
                                 mode='lines+markers', name='Area', line=dict(color='red')))

        # Update layout
        fig.update_layout(
            title=f'{crop} Year Analysis',
            xaxis_title='Year',
            yaxis_title='Normalized Values',
            legend_title='Metric',
            template='plotly_white'
        )

        # Show the plot
        fig.show()

    else:
        print(f"Data for {crop} is missing or does not contain required columns.")
```

2.To identify significant increases or decreases in crop production over the years.

```python
# List of crops
crops = ['Paddy', 'Pepper', 'Turmeric', 'Arecanut', 'Banana', 'Betel leaves', 'Cardamom',
         'Cashew', 'Cloves', 'Cocoa', 'Coconut', 'Coffee', 'Ginger', 'Jack', 'Mango',
         'Nutmeg', 'Pappaya', 'Pineapple', 'Plantain', 'Rubber', 'Tapioca', 'Tea']

# Define a threshold for significant changes (percentage)
threshold = 0.2  # For example, 20%

# Function to detect significant changes
def detect_significant_changes(df, threshold):
    df = df.sort_values(by='Agriculture Year')
    df['Production Change'] = df['Production'].pct_change()  # Calculate percentage change
    significant_changes = df[abs(df['Production Change']) > threshold]
    return significant_changes

# Iterate through each crop in the list
for crop in crops:
    # Load the dataset corresponding to the crop
    df = globals().get(crop, None)  # Use globals() to get the DataFrame with the crop name

    if df is not None and 'Agriculture Year' in df.columns and 'Production' in df.columns:
        # Detect significant changes in production
        significant_changes = detect_significant_changes(df, threshold)

        # Create a bar plot to visualize production over years
        plt.figure(figsize=(14, 7))
        sns.barplot(data=df, x='Agriculture Year', y='Production', color='skyblue', ci=None)

        # Highlight significant changes
        if not significant_changes.empty:
            sns.scatterplot(data=significant_changes, x='Agriculture Year', y='Production',
                            color='red', s=100, edgecolor='w', label='Significant Changes')

            for _, row in significant_changes.iterrows():
                plt.text(row['Agriculture Year'], row['Production'], f"{row['Production Change']:.2%}",
                         fontsize=10, ha='center', color='black')

        plt.title(f'Production Over Years for {crop}')
        plt.xlabel('Year')
        plt.ylabel('Production')
        plt.xticks(rotation=90)
        plt.legend(title='Significant Changes')
        plt.grid(axis='y', linestyle='--', alpha=0.7)  # Grid lines only on y-axis
        plt.tight_layout()
        plt.show()
    else:
        print(f"Data for {crop} is missing or does not contain required columns.")
```

##3. To compare the productivity and production of different crops.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

# List of crops
crops = ['Paddy', 'Pepper', 'Turmeric', 'Arecanut', 'Banana', 'Betel leaves', 'Cardamom',
         'Cashew', 'Cloves', 'Cocoa', 'Coconut', 'Coffee', 'Ginger', 'Jack', 'Mango',
         'Nutmeg', 'Pappaya', 'Pineapple', 'Plantain', 'Rubber', 'Tapioca', 'Tea']

# Initialize lists to collect data
crop_names = []
years = []
productions = []
productivities = []

# Collect data from all crops
for crop in crops:
    # Load the dataset corresponding to the crop
    df = globals().get(crop, None)  # Use globals() to get the DataFrame with the crop name

    if df is not None and 'Agriculture Year' in df.columns and 'Production' in df.columns and 'Productivity' in df.columns:
        for year in df['Agriculture Year'].unique():
            crop_names.append(crop)
            years.append(year)
            productions.append(df[df['Agriculture Year'] == year]['Production'].values[0])
            productivities.append(df[df['Agriculture Year'] == year]['Productivity'].values[0])

# Create a combined DataFrame
combined_df = pd.DataFrame({
    'Crop': crop_names,
    'Year': years,
    'Production': productions,
    'Productivity': productivities
})

# Calculate average production and productivity for each crop
avg_production_df = combined_df.groupby('Crop', as_index=False)['Production'].mean()
avg_productivity_df = combined_df.groupby('Crop', as_index=False)['Productivity'].mean()

# Remove crops with very low values (e.g., less than 0.1 of the max value)
production_threshold = 0.1 * avg_production_df['Production'].max()
productivity_threshold = 0.1 * avg_productivity_df['Productivity'].max()

avg_production_df = avg_production_df[avg_production_df['Production'] >= production_threshold]
avg_productivity_df = avg_productivity_df[avg_productivity_df['Productivity'] >= productivity_threshold]

# Sort the data in descending order for better visualization
avg_production_df = avg_production_df.sort_values(by='Production', ascending=False)
avg_productivity_df = avg_productivity_df.sort_values(by='Productivity', ascending=False)

# Plot the data using Seaborn
plt.figure(figsize=(14, 12))

# Production plot
plt.subplot(2, 1, 1)
sns.barplot(data=avg_production_df, y='Crop', x='Production', palette='viridis')
plt.title('Comparison of Production Across Crops')
plt.xlabel('Production (units)')
plt.ylabel('Crop')

# Productivity plot
plt.subplot(2, 1, 2)
sns.barplot(data=avg_productivity_df, y='Crop', x='Productivity', palette='viridis')
plt.title('Comparison of Productivity Across Crops')
plt.xlabel('Productivity (units)')
plt.ylabel('Crop')

plt.tight_layout()
plt.show()
```

##4. To analyze which crops are more dominant in terms of area and yield.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# List of crops
crops = ['Paddy', 'Pepper', 'Turmeric', 'Arecanut', 'Banana', 'Betel leaves', 'Cardamom',
         'Cashew', 'Cloves', 'Cocoa', 'Coconut', 'Coffee', 'Ginger', 'Jack', 'Mango',
         'Nutmeg', 'Pappaya', 'Pineapple', 'Plantain', 'Rubber', 'Tapioca', 'Tea']

# Initialize lists to collect data
crop_names = []
total_areas = []
total_yields = []

# Collect data from all crops
for crop in crops:
    # Load the dataset corresponding to the crop
    df = globals().get(crop, None)  # Use globals() to get the DataFrame with the crop name

    if df is not None and 'Area' in df.columns and 'Productivity' in df.columns:
        total_area = df['Area'].sum()
        total_yield = df['Productivity'].sum()

        crop_names.append(crop)
        total_areas.append(total_area)
        total_yields.append(total_yield)

# Create a combined DataFrame
dominance_df = pd.DataFrame({
    'Crop': crop_names,
    'Total Area': total_areas,
    'Total Yield': total_yields
})

# Determine the threshold for filtering
area_threshold = 0.1 * dominance_df['Total Area'].max()
yield_threshold = 0.1 * dominance_df['Total Yield'].max()

# Filter out crops with values below the threshold
dominance_df_filtered_by_area = dominance_df[dominance_df['Total Area'] >= area_threshold]
dominance_df_filtered_by_yield = dominance_df[dominance_df['Total Yield'] >= yield_threshold]

# Sort the filtered DataFrames
dominance_df_sorted_by_area = dominance_df_filtered_by_area.sort_values(by='Total Area', ascending=False)
dominance_df_sorted_by_yield = dominance_df_filtered_by_yield.sort_values(by='Total Yield', ascending=False)

# Plot the data
fig, axes = plt.subplots(2, 1, figsize=(25, 12))

# Total Area plot (Horizontal Bar Chart)
sns.barplot(ax=axes[0], data=dominance_df_sorted_by_area, y='Crop', x='Total Area', palette='viridis')
axes[0].set_title('Total Area Dominance Across Crops')
axes[0].set_xlabel('Total Area (units)')
axes[0].set_ylabel('Crop')

# Adding annotations on the bars for Total Area
for p in axes[0].patches:
    width = p.get_width()
    axes[0].annotate(f'{width:.2f}',
                     xy=(width, p.get_y() + p.get_height() / 2),
                     xytext=(3, 0),  # 3 points horizontal offset
                     textcoords='offset points',
                     ha='left', va='center')

# Total Yield plot (Horizontal Bar Chart)
sns.barplot(ax=axes[1], data=dominance_df_sorted_by_yield, y='Crop', x='Total Yield', palette='viridis')
axes[1].set_title('Total Yield Dominance Across Crops')
axes[1].set_xlabel('Total Yield (units)')
axes[1].set_ylabel('Crop')

# Adding annotations on the bars for Total Yield
for p in axes[1].patches:
    width = p.get_width()
    axes[1].annotate(f'{width:.2f}',
                     xy=(width, p.get_y() + p.get_height() / 2),
                     xytext=(3, 0),  # 3 points horizontal offset
                     textcoords='offset points',
                     ha='left', va='center')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# List of crops
crops = ['Paddy', 'Pepper', 'Turmeric', 'Arecanut', 'Banana', 'Betel leaves', 'Cardamom',
         'Cashew', 'Cloves', 'Cocoa', 'Coconut', 'Coffee', 'Ginger', 'Jack', 'Mango',
         'Nutmeg', 'Pappaya', 'Pineapple', 'Plantain', 'Rubber', 'Tapioca', 'Tea']

# Initialize lists to collect data
crop_names = []
average_productivities = []

# Collect data from all crops
for crop in crops:
    # Load the dataset corresponding to the crop
    df = globals().get(crop, None)  # Use globals() to get the DataFrame with the crop name

    if df is not None and 'Productivity' in df.columns:
        average_productivity = df['Productivity'].mean()

        crop_names.append(crop)
        average_productivities.append(average_productivity)

# Create a DataFrame
productivity_df = pd.DataFrame({
    'Crop': crop_names,
    'Average Productivity': average_productivities
})

# Plot the data
plt.figure(figsize=(12, 10))  # 12 inches wide and 10 inches tall

# Create pie chart
sizes = productivity_df['Average Productivity']
labels = productivity_df['Crop']
colors = sns.color_palette('viridis', len(sizes))

# Calculate explode values (slightly separate each slice)
explode = [0.05] * len(sizes)  # Adjust the value to control the separation

# Calculate percentages
total_size = sum(sizes)
percentages = [(size / total_size) * 100 for size in sizes]

# Create the legend labels with the correct percentages
legend_labels = [
    f'{label}: {size:.2f} units ({pct:.1f}%)'
    for label, size, pct in zip(labels, sizes, percentages)
]

# Plot pie chart
wedges, texts, autotexts = plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, pctdistance=0.85, explode=explode)

# Add legend to the plot
plt.legend(legend_labels, title="Crop: Productivity (units) and Percentage", loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Average Productivity Distribution Across Different Crops')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
